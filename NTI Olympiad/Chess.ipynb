```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
data = pd.read_csv('input.csv')
```

```python
win = []
```

```python
for line in data['moves']:
    win.append(1) if len(line.split(' ')) % 2 == 0 else win.append(0)
```

```python
data = pd.concat([data, pd.DataFrame({'winner' : win})], axis=1)
```

```python
data['winner'].plot.hist()
```

```python
data.columns
```

```python
data['rating_diff'] = data['white_rating'] - data['black_rating']
```

```python
data['rating_diff'].head()
```

```python
data['time'] = data['last_move_at'] - data['created_at']
```

```python
pd.DataFrame({'winner' : win}).to_csv('out.csv')
```

```python
with open('out.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['winner'])
    for w in win:
        if w == 0:
            writer.writerow(['white'])
        else:
            writer.writerow(['black'])
```
