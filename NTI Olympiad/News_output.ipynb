```python
import pandas as pd
import numpy as np
```

```python
data = pd.concat([pd.read_csv('rurur.csv'), pd.read_csv('final.csv').drop(columns=['Unnamed: 0'])], axis=0, sort=True)
test = pd.read_csv('datas/test.csv')
```

```python
data = data.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'])
data = data.drop_duplicates()
data = data.dropna()
```

```python
data
```

```python
text = "print"
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x_train = scaler.fit_transform(text)
```

```python
w = data.loc[data.theme == 'world', ['text', 'theme']][:100]
p = data.loc[data.theme == 'politics', ['text', 'theme']][:100]
e = data.loc[data.theme == 'economy', ['text', 'theme']][:100]
s = world = data.loc[data.theme == 'society', ['text', 'theme']][:100]
c = world = data.loc[data.theme == 'culture', ['text', 'theme']][:100]
r = world = data.loc[data.theme == 'religion', ['text', 'theme']][:100]
sc = world = data.loc[data.theme == 'science', ['text', 'theme']][:100]
data = pd.concat([w, p, e, s, c, r, sc], axis=0, sort=False)
```

```python
data.head()
```

```python
data.theme.value_counts()
```

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
documents = [TaggedDocument(doc, [i]) for i, doc in enumerate(data.text)]
model = Doc2Vec(documents, vector_size=80, window=2, min_count=1, workers=8)
```

```python
joja = [model.infer_vector([i]) for i in data.text]
X_train = pd.DataFrame(joja)
jojaB = [model.infer_vector([i]) for i in test.text]
X_test = pd.DataFrame(jojaB)
```

```python
y_train = data.theme
# X_train = data.drop(['theme', 'text'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, nan, y_train, nan = train_test_split(X_train, y_train, test_size=0, random_state=42)
```

```python
X_train
```

```python
pd.concat([X_train, pd.DataFrame(list(y_train))], axis=1).to_csv("out.csv")
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, max_depth=30, random_state=0)
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
```

```python
pd.DataFrame(clf.predict(X_test))[0].value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial').fit(X_train, y_train)
predictions = clf.predict(X_test)
clf.score(X_train, y_train)
```

```python
pd.DataFrame(clf.predict(X_test))[0].value_counts()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import SGDClassifier
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
text_clf = Pipeline([
('vect', CountVectorizer()),
('tfidf', TfidfTransformer()),
('clf', SGDClassifier(loss='modified_huber', penalty='l2', alpha=1e-3, random_state=42, max_iter=10, tol=None)),])
```

```python
text_clf.fit(X_train, y_train)
```

```python
predictions = text_clf.predict(X_test)
```

```python
import time
import os
import csv

current_timestamp = int(time.time())
submission_path = '{}.csv'.format(current_timestamp)
print(submission_path)

with open(submission_path, "w", newline='') as csv_file:
        writer = csv.writer(csv_file, delimiter=',')
        writer.writerow(['category'])
        for line in list(predictions):
                writer.writerow([line])
```
